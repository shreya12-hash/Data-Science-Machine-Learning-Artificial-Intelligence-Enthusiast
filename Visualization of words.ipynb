**Here we import a onlive csv file then convert it to dataframe by importing Pandas library**


```
import pandas as pd
import numpy as np

# Import clean data 
path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/module_5_auto.csv'
df = pd.read_csv(path)
```

**Now we want to find the frequency of words 'different car types' in the column 'make' by using 'groupby' and'groupby.count()'**

```
df1.drop(['Unnamed: 0.1', 'Unnamed: 0.1.1', 'symboling',
       'normalized-losses','aspiration', 'num-of-doors', 'body-style', 'drive-wheels',
       'engine-location', 'wheel-base', 'length', 'width', 'height',
       'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size',
       'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower',
       'peak-rpm', 'city-mpg', 'highway-mpg', 'price', 'city-L/100km',
       'horsepower-binned', 'diesel'],axis=1,inplace=True)
```

```
df1.head()
```

```
groupby=df1.groupby('make')
```

```
df1=groupby.count()
df1.head()
```

```
df1.drop('gas',axis=1,inplace=True)
```

```
df1.head()
```

```
df1.reset_index()
```

```
d2=df1.rename(columns={'Unnamed: 0':'Freq'}).reset_index()
```

```
d2
```

```
d3=d2.drop(20)
```

```
d3
```

```
D=d3.drop(21)
```

```
D
```

**Now merge the two dataframe D, the dataframe of different car types in column 'make' and their corresponding frequency and the original dataframe.**

```
Newdf=pd.merge(df,d3,how='left',left_index=True,right_index=True) 
```

```
Newdf.head()
```
